### Create a random numpy array with dimensions 8x2x4. *(5 points)*

```python
rand_arr = np.random.rand(8, 2, 4)
rand_arr
```

#### Must be a one-liner.

```python
identity_matrix = np.eye(8)
identity_matrix
```

### Flatten both arrays from above and sum them together element-wise. *(5 points)*

```python
rand_arr_flat = rand_arr.flatten()
identity_matrix_flat = identity_matrix.flatten()
sum_flat = rand_arr_flat + identity_matrix_flat
sum_flat
```

### Filter above flattened array for only values between 0.3 and 0.7. *(5 points)*

```python
filtered = sum_flat[(sum_flat >= 0.3) & (sum_flat <= 0.7)]
filtered
```

## Now we do some data manipulation! *(75 points)*

```python
from sklearn.datasets import load_iris

# load dataset
iris_data = load_iris()
print(iris_data.DESCR)
```

```python
# create dataframe from the dataset
df = pd.DataFrame(data=iris_data["data"], columns=iris_data["feature_names"])
df["target"] = iris_data["target"]

# map target values to species names
species_mapping = {0: "setosa", 1: "versicolor", 2: "virginica"}
df["species"] = df["target"].map(species_mapping)
df.head()
```

### Get the NumPy version of the dataset (only the first 4 columns). *(2.5 points)*
#### HINT: Check out the `.values` attribute.

```python
np_arr = df.iloc[:, :4].values
np_arr
```

### Calculate the mean of each column in the NumPy array. *(2.5 points)*
#### Must be a one-liner.

```python
mean_values = np.mean(np_arr, axis=0)
mean_values
```

### Calculate the standard deviation of each column in the NumPy array. *(2.5 points)*
#### Must be a one-liner.

```python
std_values = np.std(np_arr, axis=0)
std_values
```

### Using the above values, normalize the dataset. *(5 points)*
#### Normalizing is done by subtracting the mean and dividing by the standard deviation.
#### Must be a one-liner.

```python
normalized_arr = (np_arr - mean_values) / std_values
normalized_arr
```

### Filter the dataset to include only rows with petal length strictly greater than the average petal length. *(5 points)*  

```python
average_petal_length = df["petal length (cm)"].mean()
filtered_rows = df[df["petal length (cm)"] > average_petal_length]
filtered_rows
```

### Create a new column for calculating `sepal area`–just use the standard formula for the area of a rectangle. *(5 points)*

```python
df["sepal_area"] = df["sepal length (cm)"] * df["sepal width (cm)"]
```

### Compute the `median` and the `mean` of the new column. *(5 points)*

```python
median_sepal_area = df.groupby("species")["sepal_area"].median()
mean_sepal_area = df.groupby("species")["sepal_area"].mean()

median_sepal_area, mean_sepal_area
```

### Filter the dataframe for the 5 smallest sepal areas. *(5 points)*
#### Do it without sorting.

```python
smallest_sepal_areas = df.nsmallest(5, "sepal_area")
smallest_sepal_areas
```

### Find the min/max values for each feature by species. *(10 points)*

```python
min_max_values = df.groupby("species").agg(["min", "max"])
min_max_values
```

### Normalize the dataframe. (subtract the mean and divide by the standard deviation)
### Also, sort by sepal length in descending order. *(10 points)*
#### Make sure to copy the dataframe first.

```python
normalized_df = df.copy()[
    ["sepal length (cm)", "sepal width (cm)", "petal length (cm)", "petal width (cm)"]
]

normalized_df = (normalized_df - normalized_df.mean()) / normalized_df.std()
normalized_df.sort_values("sepal length (cm)", ascending=False)
normalized_df.head()
```

## Data Visualization *(20 points)*

Make sure to display the plots with titles, axis labels, etc. Just in general, make them look nice!

### Plot Sepal Length vs. Sepal Width. *(10 points)*
#### Make sure to have a legend for the species.

```python
species_colors = {"setosa": "red", "versicolor": "green", "virginica": "blue"}
plt.figure(figsize=(8, 6))

for species in df["species"].unique():
    species_data = df[df["species"] == species]
    plt.scatter(
        species_data["sepal length (cm)"],
        species_data["sepal width (cm)"],
        color=species_colors[species],
        label=species,
    )

plt.title("Sepal Length vs Sepal Width by Species")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm)")
plt.legend(title="Species")
plt.show()
```

### Plot a histogram of the petal length. *(10 points)*
#### Use `10` bins.
#### Make sure to again include the species for the legend.
#### There may be some overlap in the data, so make sure to include some transparency in the plot. (`alpha` parameter)

```python
# Histogram of petal length
plt.figure(figsize=(8, 6))

for species in df["species"].unique():
    species_data = df[df["species"] == species]
    plt.hist(
        species_data["petal length (cm)"],
        bins=10,
        alpha=0.5,
        label=species,
        color=species_colors[species],
    )

plt.title("Petal Length Distribution by Species")
plt.xlabel("Petal Length (cm)")
plt.ylabel("Frequency")
plt.legend(title="Species")
plt.show()
```
